1) 10 Player terbanyak mencetak goal

```python
top_goal_scorers = df.sort_values(by='goals', ascending=False).head(10)
print("10 Player terbanyak mencetak gol:")
top_goal_scorers[['player_name', 'goals']]
```

2)10 Player terlama durasi bermain

```python
top_play_time = df.sort_values(by='time', ascending=False).head(10)
print("10 Player terlama durasi bermain:")
top_play_time[['player_name', 'time']]
```

3)10 Player terbanyak assist

```python
top_assisters = df.sort_values(by='assists', ascending=False).head(10)
print("10 Player terbanyak assist:")
top_assisters[['player_name', 'assists']]
```

4) 4 player pencetak gol terbanyak dari kesebelasan manchaster city

```python
mc_top_goal_scorers =df[df['team_title'] == 'Manchester City'].sort_values(by='goals', ascending=False).head(4)
print("4 Player pencetak gol terbanyak dari kesebelasan Manchester City:")
mc_top_goal_scorers[['player_name', 'goals']]
```

5) Dari kesebelasan mana pencetak gol terbanyak

```python
most_goals_team = df.sort_values(by='goals', ascending=False).iloc[0]
print("Dari kesebelasan mana pencetak gol terbanyak:")
most_goals_team[['team_title']]
```

6) Berapa jumlah kartu kuning dan merah yang diterima pembuat assist terbanyak

```python
top_assist_player = df.sort_values(by='assists', ascending=False).iloc[0]
print("Jumlah kartu kuning dan merah yang diterima pembuat assist terbanyak:")
top_assist_player[['player_name', 'yellow_cards','red_cards']]
```

7) Siapa yang jumlah assist dan golnya paling banyak dan dari kesebelasan mana?

```python
most_goals_assists_player = df.loc[(df['goals'] + df['assists']).idxmax()]
team_most_goals_assists_player = most_goals_assists_player['team_title']
print("Pemain dengan jumlah assist dan gol terbanyak dan dari kesebelasan mana:")
most_goals_assists_player[['player_name', 'goals','assists','team_title']]
```

8) 5 kesebelasan yang menyumbangkan gol terbanyak

```python
top_5_goal_contributors = df.groupby('team_title')['goals'].sum().nlargest(5)
print("5 Kesebelasan yang menyumbangkan gol terbanyak:")
top_5_goal_contributors
```

9) Urutkan pemain berdasarkan jumlah gol terbanyak tanpa mendapatkan kartu sama sekali

```python
players_no_cards = df[(df['yellow_cards'] == 0) & (df['red_cards'] == 0)].sort_values(by='goals', ascending=False)
print("Pemain yang memiliki jumlah gol terbanyak tanpa mendapatkan kartu sama sekali:")
players_no_cards[['player_name', 'goals']]
```
